# EDA for IoT Telemetry Data
This notebook performs exploratory data analysis on the synthetic IoT dataset.

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_path = '../data/iot_telemetry.csv'
if not os.path.exists(data_path):
    data_path = 'data/iot_telemetry.csv'

df = pd.read_csv(data_path)
df.head()
```

```python
df.info()
```

```python
df['is_anomaly'].value_counts(normalize=True).plot(kind='bar')
plt.title('Target Distribution: Anomaly vs Normal')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
